```python
import duckdb

# Connect to the DuckDB database√ü
conn = duckdb.connect('taxi_rides_ny/taxi_rides_ny.duckdb')
```

## Question 3: Count of records in fct_monthly_zone_revenue

```python
# Question 3: Count records in fct_monthly_zone_revenue
query = """
SELECT COUNT(*) as record_count
FROM prod.fct_monthly_zone_revenue
"""
result = conn.execute(query).fetchall()
print(f"Total records in fct_monthly_zone_revenue: {result[0][0]:,}")
```

## Question 4: Best Performing Zone for Green Taxis (2020)

```python
# Question 4: Highest revenue zone for Green taxis in 2020
query = """
SELECT 
    pickup_zone,
    SUM(revenue_monthly_total_amount) as total_revenue
FROM prod.fct_monthly_zone_revenue
WHERE service_type = 'Green'
    AND EXTRACT(YEAR FROM revenue_month) = 2020
GROUP BY pickup_zone
ORDER BY total_revenue DESC
LIMIT 10
"""
result = conn.execute(query).fetchall()
print("Top 10 zones by revenue for Green taxis in 2020:")
for row in result:
    print(f"{row[0]}: ${row[1]:,.2f}")
```

## Question 5: Green Taxi Trip Counts (October 2019)

```python
# Question 5: Total trips for Green taxis in October 2019
query = """
SELECT 
    SUM(total_monthly_trips) as total_trips
FROM prod.fct_monthly_zone_revenue
WHERE service_type = 'Green'
    AND revenue_month = '2019-10-01'
"""
result = conn.execute(query).fetchall()
print(f"Total Green taxi trips in October 2019: {result[0][0]:,}")
```
